```python
%%writefile modules/consts.py
''' Descripción:
        
'''
from pysonance.const import *
from pysonance.control import *
from pysonance.filter import *
from pysonance.IO import *
from pysonance.sampler import *
from pysonance.signal import *
from pysonance.sound import *

reg = Register_IO('pactl')

## CODIDO DEL MÓDULO AQUÍ
valores = [1, 0.5, 0.3, 2]
for v in valores:
    _c = C(v)
    _lo = Line_Out(f'c2_{v}', _c) # generamos salida para cada constante
    reg.add(_lo)

## Mantener el módulo en ejecución
nombre = 'consts'
print(f'{nombre}:')
reg.start()
print('Enter para parar')
input()  # mantener el módulo en ejecución hasta que se presione Enter
reg.stop()
print('Parado')
```

```python
%%writefile modules/echo.py
''' Descripción:
        
'''
from pysonance.const import *
from pysonance.control import *
from pysonance.filter import *
from pysonance.IO import *
from pysonance.sampler import *
from pysonance.signal import *
from pysonance.sound import *

reg = Register_IO('pactl')

## CODIDO DEL MÓDULO AQUÍ
# cada vez que crees un modulo, has de agregarlo al registro (reg.add)

i = I('echo_in', default=Sine(440, amp=0.5))
o = O('echo_out', i)

reg.add(i)
reg.add(o)

## Mantener el módulo en ejecución
nombre = 'plantilla'
print(f'{nombre}:')
reg.start()
print('Enter para parar')
input()  # mantener el módulo en ejecución hasta que se presione Enter
reg.stop()
print('Parado')
```
